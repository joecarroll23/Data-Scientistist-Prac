```python
data = load_iris()
```

```python
iris_df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
iris_df['species'] = data.target_names[data.target]
```

```python
iris_df
```

```python

plt.figure(figsize=(8, 6))
for species, color in zip(data.target_names, ['blue', 'green', 'red']):
    plt.scatter(iris_df[iris_df['species'] == species]['sepal length (cm)'],
                iris_df[iris_df['species'] == species]['sepal width (cm)'],
                c=color, label=species, alpha=0.8, edgecolors='w')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot: Sepal Length vs. Sepal Width')
plt.legend()
plt.grid(True)
sepal_len_v_width_sp = """
 Setosa has shorter sepal length and wider sepal width
Versicolor has moderate sepal length and sepal width
Virginica  has longer sepal lengths but thinder speal widths
There is some overlapping between Versicolor and Virginica
"""
plt.text(10, 3, sepal_len_v_width_sp, fontsize=10, bbox=dict(facecolor='lightgray', alpha=0.5))

plt.show()
```

```python
plt.figure(figsize=(8, 6))
for species, color in zip(data.target_names, ['blue', 'green', 'red']):
    plt.scatter(iris_df[iris_df['species'] == species]['petal length (cm)'],
                iris_df[iris_df['species'] == species]['petal width (cm)'],
                c=color, label=species, alpha=0.8, edgecolors='w')
plt.xlabel('Petal Length in cm')
plt.ylabel('Petal Width in cm')
plt.title('Scatter Plot of  Petal Length and Petal Width')
plt.legend()
plt.grid(True)
petal_len_v_width_sp = """
Setosa has shortest width and length but their ratio is the highest
Versicolor is moderate in terms of width of length
Virginica has the longest petal and is also widest

There is minimal overlap in this model and each group is clearly defined
"""
plt.text(10, 3, sepal_len_v_width_sp, fontsize=10, bbox=dict(facecolor='lightgray', alpha=0.5))

plt.show()
```
